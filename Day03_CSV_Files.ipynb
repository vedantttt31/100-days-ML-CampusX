<h2>3. Opening a CSV file from a URL<h2/>

```python
import requests
from io import StringIO

url = ""
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

<h2>4. sep parameter<h2/>

```python
# For tab-separated files
df = pd.read_csv('file.csv', sep='\t')

# For semicolon-separated files
df = pd.read_csv('file.csv', sep=';')

# For pipe-separated files
df = pd.read_csv('file.csv', sep='|')
```

<h2>5. index_col parameter (to set a column as the index)<h2/>

```python
# Set first column as index
df = pd.read_csv('aug_train.csv', index_col=0)

# Set a specific column by name as index
df = pd.read_csv('aug_train.csv', index_col='enrollee_id')

# Set multiple columns as index
df = pd.read_csv('aug_train.csv', index_col=[0, 1])
```

<h2>6. Header parameter<h2/>

```python
# Default (first row as header)
pd.read_csv('test.csv')

# No header in file
pd.read_csv('test.csv', header=None)

# Use 2nd row as header
pd.read_csv('test.csv', header=1)
```

<h2>7. usecols parameter<h2/>

```python
# Select specific columns by name
pd.read_csv('aug_train.csv', usecols=['enrollee_id', 'city', 'target'])

# Select specific columns by index
pd.read_csv('aug_train.csv', usecols=[0, 1, 2])
```

<h2>8. Squeeze parameter<h2/>

```python
# Returns a Series instead of DataFrame if only one column
pd.read_csv('aug_train.csv', usecols=['enrollee_id'], squeeze=True)

# Note: squeeze is deprecated in newer pandas, use this instead
pd.read_csv('aug_train.csv', usecols=['enrollee_id']).squeeze()
```

## 9. Skiprows/nrows Parameter

```python
pd.read_csv('test.csv', skiprows=[1, 2])
pd.read_csv('test.csv', nrows = 100) # nrows restric the total number of rows to say 100 if their are more than that rows
```

## 10. Encoding parameter

```python
# Default encoding
pd.read_csv('zomato.csv')

# With encoding specified
pd.read_csv('zomato.csv', encoding='utf-8')

# For files with special characters
pd.read_csv('zomato.csv', encoding='latin-1')

# Another common encoding
pd.read_csv('zomato.csv', encoding='iso-8859-1')
```

## 11. Skip bad lines

```python
# Skip bad/corrupted lines
pd.read_csv('BX-Books.csv', sep=';', encoding='latin-1', on_bad_lines='skip')

# Older pandas version used this instead
pd.read_csv('BX-Books.csv', sep=';', encoding='latin-1', error_bad_lines=False)
```

## 12. dtypes parameter

```python
# Specify data types for columns
pd.read_csv('aug_train.csv', dtype={'enrollee_id': int, 'city': str, 'target': float})

# Set all columns as string
pd.read_csv('aug_train.csv', dtype=str)

# Set specific column dtype
pd.read_csv('aug_train.csv', dtype={'target': 'int32'})
```

## 13. Handling Dates

```python
# Parse a date column automatically
pd.read_csv('file.csv', parse_dates=['date_column'])

# Parse multiple date columns
pd.read_csv('file.csv', parse_dates=['start_date', 'end_date'])

# Combine multiple columns into one date
pd.read_csv('file.csv', parse_dates={'date': ['year', 'month', 'day']})
```

## 14. Loading a huge dataset in chunks

```python
# Load dataset in chunks of 5000 rows
dfs = pd.read_csv('aug_train.csv', chunksize=5000)

# Iterate through chunks
for chunk in dfs:
    print(chunk.shape)


##Bonus
# Concatenate all chunks into one DataFrame
import pandas as pd

dfs = pd.read_csv('aug_train.csv', chunksize=5000)
df = pd.concat(dfs, ignore_index=True)

# Process each chunk and collect results
results = []
for chunk in pd.read_csv('aug_train.csv', chunksize=5000):
    # do some operation on each chunk
    results.append(chunk[chunk['target'] == 1])

final_df = pd.concat(results, ignore_index=True)
```
